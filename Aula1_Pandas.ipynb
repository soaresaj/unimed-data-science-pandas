**Análise de dados com Python e Pandas - Introdução à biblioteca Pandas**

```python
# Importando a biblioteca Pandas:
import pandas as pd # para facilitar no código é atribuído um nome reduzido para a biblioteca "pd"
```

```python
# Vamos importar uma base de dados que está no formato ".csv" gravada no meu google drive para um dataframe "df"
# vou ler utilizando o parâmetro "error_bad_lines" para ignorar possíveis linhas com erro no arquivo "Gapminder.csv":
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data/Gapminder.csv", error_bad_lines=False)
```

```python
# Visualizando as 5 primeiras linhas do dataframe "df":
df.head()
```

```python
# O formato da visualização está estranho, percebemos que o separador de campos no arquivo ".csv" é ";"
# ao invés do  default ",".
# Então vamos colocar mais um parâmetro para a leitura do database para a criação do dataframe:
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data/Gapminder.csv", error_bad_lines=False, sep=";")
```

```python
# Visualizando novamente as 5 primeiras linhas do dataframe "df", agora os dados são apresentados como uma tabela:
df.head()
```

```python
# Agora vamos utilizar o método "rename" para alterar o nome das colunas:
df.rename(columns={'country':'Pais', 'continent':'Continente', 'year':'Ano', 'lifeExp':'Expectativa de Vida', 'pop':'Pop Total', 'gdpPercap':'PIB'})
```

```python
# Para que a alteração esteja no meu dataframe tenho que atribuir o resultado a variável "df":
df = df.rename(columns={'country':'Pais', 'continent':'Continente', 'year':'Ano', 'lifeExp':'Expectativa de Vida', 'pop':'Pop Total', 'gdpPercap':'PIB'})
```

```python
df.head(10)
```

```python
# Quantas linhas e colunas tenho no dataframe? Vamos utilizar a propriedade "shape" do mesmo:
df.shape
```

```python
# Com a propriedade "columns" consigo verificar os nomes das colunas do dataframe:
df.columns
```

```python
# Agora quero saber o tipo de dado de cada coluna do dataframe, vou utilizar a propriedade "dtypes":
df.dtypes
```

```python
# Como faço para verificar as últimas linhas do dataframe, uso o método "tail":
df.tail(10)
```

```python
# Já o método "describe" apresenta informações estatísticas básicas para cada uma das colunas do dataframe
# ou para uma coluna específica:
df.describe()
```

```python
# Posso fazer filtros no dataframe, utilizando o nome das colunas e métodos como o "unique" que apresenta apenas 
# os valores únicos da  coluna selecionada:
df['Continente'].unique()
```

```python
# Posso filtrar um valor específico no dataframe utilizando o método "loc":
oceania = df.loc[df['Continente'] == 'Oceania']
oceania.head()
```

```python
# Se quiser granteir que no novo dataframe tenha apenas os dados do continente oceania,
# utilizo novamente o método "unique"
oceania['Continente'].unique()
```

```python
# Se quero agrupar as informações no dataframe posso utilizar o método "groupby":
df.groupby("Continente")["Pais"].nunique() # agrupo por continente e verifico quantos países em cada continente com o método "nunique"
```

```python
# Utilizando novamente o método "groupby" posso apresentar a expectativa de vida média em cada "Ano" no dataframe:
df.groupby("Ano")["Expectativa de Vida"].mean() # utilizo o método "mean" para apresentar a média da expectiva por ano
```

```python
# Posso utilizar o método "mean" diretamente numa coluna do dataframe:
df["PIB"].mean() # apresenta o 'PIB' médio de todos os países considerando todos as informações anuais de cada um
```

```python
# Posso utilizar o método "sum" diretamente numa coluna do dataframe:
df["PIB"].sum() # apresenta o 'PIB' total de todos os países considerando todos as informações anuais de cada um
```
